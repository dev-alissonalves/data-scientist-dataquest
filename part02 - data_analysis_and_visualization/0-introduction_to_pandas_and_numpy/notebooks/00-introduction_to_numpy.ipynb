* NumPy , abreviação de "Python Numérico", é uma biblioteca fundamental para computação científica em Python. É um favorito entre os programadores porque simplifica tarefas complexas. Como Python é uma linguagem de alto nível, não precisamos nos preocupar em alocar memória manualmente. As linguagens de baixo nível, por outro lado, exigem que definamos a alocação e o processamento de memória, o que oferece mais controle, mas pode tornar nossa programação mais lenta. NumPy atinge o equilíbrio perfeito: processamento rápido sem o incômodo de alocação manual.

```python
# NYC Taxi-Airport Data

import csv
import numpy as np

# import nyc_taxi.csv as a list of lists
f = open("../dataset/nyc_taxis.csv", "r")
taxi_list = list(csv.reader(f))

# remove the header row
taxi_list = taxi_list[1:]

# convert all values to floats
converted_taxi_list = []
for row in taxi_list:
    converted_row = []
    for element in row:
        converted_row.append(float(element))
    converted_taxi_list.append(converted_row)

# start writing your code below this comment
taxi = np.array(converted_taxi_list)
print(taxi[:5])
```
