- **Visualizao exploratria de dados**: criamos grficos para melhor compreender e explorar os dados.

- **Visualizao explicativa de dados**: criamos grficos para que outras pessoas possam informar, defender uma ideia ou contar uma histria.




```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../dataset/top20_deathtoll.csv')
top20_deathtoll = data.head(20)

# Criar o grfico
plt.barh(data['Country_Other'], data['Total_Deaths'], color='skyblue')
plt.xlabel('number of total deaths')
plt.ylabel('the top 20 countries')
plt.title('SARS-COV2 2020')

# Mostrar o grfico
plt.show()
```
