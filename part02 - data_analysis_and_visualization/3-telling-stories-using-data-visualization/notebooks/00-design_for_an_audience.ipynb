```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../dataset/top20_deathtoll.csv')
top20_deathtoll = data.head(20)

# Criar o grfico
plt.barh(data['Country_Other'], data['Total_Deaths'], color='skyblue')
plt.xlabel('number of total deaths')
plt.ylabel('the top 20 countries')
plt.title('SARS-COV2 2020')

# Mostrar o grfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../dataset/top20_deathtoll.csv')
top20_deathtoll = data.head(20)

# Criar o grfico
fig, ax = plt.subplots()
ax.barh(data['Country_Other'], data['Total_Deaths'])

# Mostrar o grfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'])

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(['A', 'B', 'C'],
       [2, 4, 16])
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(['A', 'B', 'C'],
       [2, 4, 16])

for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(['A', 'B', 'C'],
       [2, 4, 16])

ax.tick_params(bottom=False, left=False)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'])

ax.tick_params(bottom = False, left = False)

for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)

plt.show()
```

```python
# Erasing Redundant Data-Ink

fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'], height = 0.45)

ax.tick_params(bottom = False, left = False)
ax.set_xticks([0, 100000, 200000, 300000])
for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)

plt.show()
```

```python
# The Direction of Reading

fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'],
        height=0.45, 
        color='#af0b1e')

for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)
    
ax.set_xticks([0, 150000, 300000])
ax.xaxis.tick_top()
ax.tick_params(top=False, left=False)
ax.tick_params(axis='x', colors='grey')
ax.set_xticklabels(['0', '150,000', '300,000'])
ax.set_yticklabels([])
ax.axvline(x=150000)
ax.axvline(x=150000, ymin=0.045, c='grey', alpha=0.5)
country_names = top20_deathtoll['Country_Other']
for i, country in zip(range(20), country_names):
    ax.text(x=-80000, y=i-0.15, s=country)

ax.text(x = -80000, 
        y = 23.5,
        s = 'The Death Toll Worldwide Is 1.5M+', 
        size = 17,
        weight = 'bold')

ax.text(x = -80000, 
        y = 22.5, 
        s = 'Top 20 countries by death toll (December 2020)', 
        size = 12)
plt.show()
```
