##### Gestalt Principles



- Se você vir um quadrado e um triângulo, perceberá um padrão. Se dividirmos a imagem em componentes individuais, entretanto, teremos apenas um monte de pequenos círculos cinza espaçados e organizados arbitrariamente. Então por que vemos um padrão?

- O padrão que vemos é resultado de como nosso sistema visual processa os componentes individuais. Nosso interesse como cientistas de dados é prever que tipo de padrões as pessoas veem em nossas visualizações de dados. Para prever isso, podemos usar os princípios da Gestalt .

- A ideia por trás dos princípios da Gestalt é que os humanos geralmente percebem padrões em vez de objetos individuais. Do ponto de vista prático, os princípios da Gestalt nos dizem que tipo de padrão podemos esperar que as pessoas vejam quando lhes mostramos nossas visualizações de dados. Vamos começar!
