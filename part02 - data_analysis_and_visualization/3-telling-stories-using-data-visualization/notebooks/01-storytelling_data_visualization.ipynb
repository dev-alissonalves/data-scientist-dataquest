##### Data Stories

Num sentido amplo, uma história é uma sequência de eventos: algo acontece, depois acontece outra coisa, e assim por diante. Em vez de eventos, nosso gráfico acima apresenta fatos.

Outro elemento da história é a mudança: algo ou alguém muda ao longo da história. Nosso gráfico é estático: a informação que apresentamos não apresenta nenhum elemento de mudança.



```python
# Grid Charts in Matplotlib

import pandas as pd
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows = 4, ncols = 1, figsize = (6, 8))
ax1.plot([2, 4, 8], [2, 8, 16])
ax3.plot([3, 9, 18], [16, 4, 16])

plt.show()
```
