```python
# Introduction

from csv import reader, writer
opened_file = open('../dataset/potus_visitors_2015.csv')
read_file = reader(opened_file)
potus = list(read_file)
potus = potus[1:]

print(potus[:5])
```

```python
# Importing Modules

import datetime as dt
```

```python
# The Datetime Class
# datetime.datetime(year, month, day, hour=0, minute=0, second=0)
# eg_2 = dt.datetime(1985, 3, 13, 21, 26, 2)

ibm_founded = dt.datetime(1911, 6, 16)
man_on_moon = dt.datetime(1969, 7, 20, 20, 17)

print(ibm_founded)
print(man_on_moon)
```

```python
# Using Strptime to Parse Strings as Dates - Part 1

date_1_str = "24/12/1984"
date_1_dt = dt.datetime.strptime(date_1_str, "%d/%m/%Y")
print(type(date_1_dt))
print(date_1_dt)
    

    
```

```python
# Using Strptime to Parse Strings as Dates - Part 2
date_format = '%m/%d/%y %H:%M'

for row in potus:
    date = row[2]
    row[2] = dt.datetime.strptime(date, date_format)
    print(row[2])
    
```

```python
# Using Strftime to Format Dates

visitors_per_month = {}
for row in potus:
    date = row[2]
    date = date.strftime("%B, %Y")
    if date not in visitors_per_month:
        visitors_per_month[date] = 1
    else:
        visitors_per_month[date] += 1

print(visitors_per_month)
```

### RESUMO

#### strptime: Converte uma string em um objeto datetime usando um formato especfico.
#### strftime: Formata um objeto datetime como uma string usando um formato especfico.

```python
# The Time Class - Part1
# EXAMPLES

two_thirty = dt.time(14, 30)
print(two_thirty)

five_sec_after_8am = dt.time(8,0,5)
print(five_sec_after_8am)

jfk_shot_dt = dt.datetime(1963, 11, 22, 12, 30)
print(type(jfk_shot_dt))

jfk_shot_t = jfk_shot_dt.time()
print(jfk_shot_t)
```

```python
# The Time Class - Part2

appt_times = []

for row in potus:
    datetime_potus = row[2]
    time_potus = datetime_potus.time()
    appt_times.append(time_potus)
    
```

```python
# Comparing Time Objects

# Assign the earliest appointment time from the appt_times list to the variable min_time
min_time = appt_times[0]

# Assign the latest appointment time from the appt_times list to the variable max_time
max_time = appt_times[-1]
```

```python
# Calculations with Dates and Times - part1

dt.timedelta(days=0, seconds=0, microseconds=0, milliseconds=0, minutes=0, hours=0, weeks=0)

## adding a week
d1 = dt.date(1963, 2, 21)
d1_plus_1wk = d1 + dt.timedelta(weeks=1)
print(d1_plus_1wk)
```

```python
# Calculations with Dates and Times - part2
dt1 = dt.datetime(2022, 4, 14)
dt2 = dt.datetime(2022, 3, 29)
dt3 = dt.datetime(2024, 1, 26)
dt4 = dt.datetime(2024, 1, 28)

# Calcule o tempo entre dt_2e dt_1e atribua o resultado a answer_1
answer_1 = dt2 - dt1
print(answer_1)

# Add 56 days to dt_3, and assign the result to answer_2
answer_2 = dt3 + dt.timedelta(days = 56)
print(answer_2)

# Subtract 3600 seconds from dt_4, and assign the result to answer_3
answer_3 = dt4 - dt.timedelta(seconds = 3600)
print(answer_3)
```

```python
# Summarizing Appointment Lengths - Example

start_time = dt.datetime(2015, 1, 6, 9, 30)
end_time = dt.datetime(2015, 1, 6, 10, 30)
meeting_length = end_time - start_time
print(meeting_length)

```

```python
# Summarizing Appointment Lengths

appt_lengths = {}

for row in potus:
    date_start = row[2] # Column 2 has already been converted to datetime in previous exercises
    date_end = row[3]
    date_end = dt.datetime.strptime(date_end, "%m/%d/%y %H:%M")
    row[3] = date_end
    length = date_end - date_start
    
    if length not in appt_lengths:
        appt_lengths[length] = 1
    else:
        appt_lengths[length] += 1
    
min_length = min(appt_lengths)
max_length = max(appt_lengths)

print(min_length, max_length)
```
