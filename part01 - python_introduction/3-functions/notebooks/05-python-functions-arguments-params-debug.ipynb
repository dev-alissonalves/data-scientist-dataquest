```python
# Extract Values from Any Column 

from csv import reader
opened_file = open('../dataset/AppleStore.csv')
read_file = reader(opened_file)
apps_data = list(read_file)

values_dict = {}

def extract(index_value):
    values_column = []
    for row in apps_data[1:]:
        values_column.append(row[index_value])
    
    return values_column

def freq_table(list_values):

    for elem in list_values:
        if elem in values_dict:
            values_dict[elem] += 1
        else:
            values_dict[elem] = 1
    return values_dict

genres_ft = freq_table(extract(11))

print(genres_ft)
```

```python
# Combining Functions

from csv import reader
opened_file = open('../dataset/AppleStore.csv')
read_file = reader(opened_file)
apps_data = list(read_file)
data_set = apps_data[1:]

def mean(data_set, index): 
    column = extract(data_set, index)
    return find_sum(column) / find_length(column)

def extract(data_set, index):
    column = []    
    for row in data_set:
        value = row[index]
        column.append(value)    
    return column

def find_sum(a_list):
    a_sum = 0
    for element in a_list:
        a_sum += float(element)
    return a_sum

def find_length(a_list):
    length = 0
    for element in a_list:
        length += 1
    return length


avg_price = mean(data_set, 4)
avg_rating = mean(data_set, 7)

print(f'MEAN: {avg_price:.2f}')
print(f'MEAN: {avg_rating:.2f}')
```
